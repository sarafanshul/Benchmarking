```python
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
import math
import matplotlib as mp
import random
```

```python
with open('benchmarks_data.json' ,'r') as j_file:
    data = json.load(j_file)
    names = data.keys()
    values = [sum(data[i])/len(data[i]) for i in names]
df = pd.DataFrame(data)
```

```python
plt.style.use("seaborn")#dark_background
fig = plt.figure(figsize = (18 ,10) ,
           dpi = 80,)
fig.tight_layout()

scr = [random.randint(20,100) for i in range(len(names))]
likeability_scores = np.array(scr)
data_normalizer = mp.colors.Normalize()
color_map = mp.colors.LinearSegmentedColormap(
    "my_map",{"red": [(0, 0.1, 1.0),
                    (1.0, 0.5, .7)],
             "green": [(0, 0.4, 0.9),
                      (1.0, 0.2, 0.7)],
             "blue": [(0, 0.2, 0.3),
                     (1.0, 0.7, 0.1)]}
            )

plt.bar(names,values ,
        label = "Languages",
        align="center",
        color=color_map(data_normalizer(likeability_scores))) # use log graph

plt.xlabel("Languages")
plt.ylabel("log of mean Runtime for N = 1e8")
plt.yscale('log')
plt.legend()
plt.xticks(rotation=45)
plt.subplots_adjust(bottom=0.15)

my_xtics = ("C","C++","Python","Java","Rust","Py-C","Py-Cpp","Py-Rs")
my_xtics = [my_xtics[i]+" ("+ str(values[i])[:5] + "secs)" for i in range(len(my_xtics))]
plt.xticks(list(names),my_xtics)
plt.title('Comparision between different languages w.r.t Time (Lower is Better)\n[N-Queens]',
         fontdict = {'fontsize': 18})
```
