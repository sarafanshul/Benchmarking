```python
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
import matplotlib as mp
import math
import random
%matplotlib inline
```

```python
with open('benchmarks.json' ,'r') as j_file:
    data = json.load(j_file)
df = pd.DataFrame(data)
# plt.bar(df)
names = df.keys()
# values = [math.log(sum(df[a])/len(df[a]) ,10) for a in df.keys()]
values = [sum(df[a])/len(df[a]) for a in df.keys()]
```

```python
plt.rcParams.update(plt.rcParamsDefault)
```

```python
plt.style.use("seaborn")#dark_background
fig = plt.figure(figsize = (18 ,10) ,
           dpi = 80,
# #            edgecolor = "y",
#            facecolor = "y",
          )
fig.tight_layout()
# plt.rcParams.update({'text.color' : "white",
#                      'axes.labelcolor' : "blue"})
scr = [random.randint(20,100) for i in range(len(names))]
likeability_scores = np.array(scr)
data_normalizer = mp.colors.Normalize()
color_map = mp.colors.LinearSegmentedColormap(
    "my_map",{"red": [(0, 1.0, 1.0),
                    (1.0, 0.6, .5)],
             "green": [(0, 1.0, 0.5),
                      (1.0, 0.2, 0)],
             "blue": [(0, 0.59, 0.5),
                     (1.0, 0.6, 0)]}
            )

plt.bar(names,values ,
        label = "Languages",
        align="center",
        color=color_map(data_normalizer(likeability_scores))) # use log graph

plt.xlabel("Languages")
# plt.xticks(names, ('Python-Rust Ext', 'Python-C++ Ext', 'Python', 'Cython', 'Python-C Ext' ,'C++' ,'Java' ,'Rust' ,'C'))
plt.ylabel("log of mean Runtime for N = 1e8")
plt.yscale('log')
plt.legend()
plt.xticks(rotation=45)
plt.subplots_adjust(bottom=0.15)

# axes2 = plt.twinx()
# axes2.plot(names, values, color='k', 
#            label='differences comparision')

my_xtics = ('Python-Rust Ext', 'Python-C++ Ext', 'Python', 'Cython', 'Python-C Ext' ,'C++' ,'Java' ,'Rust' ,'C')
my_xtics = [my_xtics[i]+" ("+ str(values[i])[:5] + "secs)" for i in range(len(my_xtics))]
plt.xticks(names,my_xtics)
# axes2.legend(loc = 2)
# axes2.yscale('log')
plt.yscale('log')
plt.title('Comparision between different languages w.r.t Time (Lower is Better)\nSieve of Eratosthenes',
         fontdict = {'fontsize': 18})
# plt.savefig('graph.png') # saved already
```
